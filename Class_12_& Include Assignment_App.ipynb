 *1. Count Down Recursion Function

 Def Count_Down():

   Count_Down()

 Count_Down() : GIAIC Function


```python
print("_** COUNT DOWN **_")

def countdown(giaic: int)-> int:
    print("GIAIC>>>",giaic)
    if giaic > 1:
        countdown(giaic - 1)

countdown(10)
print("Ramazan Mubarak in GIAIC-AI - Done")
```

```python
def count_down(giaic: int)-> int:
    if giaic == 55: # 0 => 0
        print("Count Down GIAIC")
    else:
        print("Giaic>>>",giaic)
        count_down(giaic + 1)


count_down(0)

print("Ramazan Mubarak in GIAIC-AI - Count Down Done")
```

 ** Class Assignment **

 ** Factorial Function =  num * num-1

```python
def factorial(n: int, giaic: int):
    if n <= 0:
        print("Fictorial of", giaic,"is:")
        return 1
    else:
        print("GIAIC",n)
    return n * factorial(n-1, giaic)
result = factorial(9, 44)
print(result)
print("Ramazan Mubarak in GIAIC-AI - Factorial Done")
```

 ** LAMBDA FUNCTION **

 *2.Lambda Parameters : Expression / statement

```python
def add(a:int,b:int) -> int:
    return a + b

result = add(5,6) # 11

print("result>>>",result)
```

 ** Class Assignment **

 ** Map Square Lambada Function **


```python
giaics = [1,2,3,4,5]
#        [1, 4, 9, 16, 25]

def lambada(giaic: int) :
    return giaic * giaic

giaic_list =list(map(lambada,giaics))
print("GIAICS",giaics)
print("GIAIC_list",giaic_list)
```

```python
giaics = [1,2,3,4,5]
#        [1, 4, 9, 16, 25]

def Lambada(giaic: int) :
    return giaic * giaic

giaic_list = []
print("GIAIC_List Before For Loop",giaic_list)
for giaic in giaics:
    print("GIAIC>>>",giaic)
    print("Calling Lambada Function")
    giaic_lamba = lambada(giaic)
    print("GIAIC_Lamba>>>",giaic_lamba)
    print("Adding Element To List")
    giaic_list.append(giaic_lamba)
    print(lambada(2))

print("GIAIC_List After For Loop",giaic_list)
```
