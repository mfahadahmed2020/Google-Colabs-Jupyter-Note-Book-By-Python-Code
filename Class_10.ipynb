SETüìê METHODS

1. Add
2. Remove
#  If The Value is Not in Set So Remove Show Error
#  But Discard & Remove & Pop Have Same Work
3. Discard
#  If The Value is Not in Set So Discard Don't Show Error
4. Pop
5. Union
6. Intersection

```python
name_set = {"A", "B", "C", "D", "E"}

# 1. Add
name_set.add("F")
print(f"Add Set: {name_set}")

# 2.Remove
name_set.remove("C")
print(f"Remove Set: {name_set}")



# 3. Discard
name_set.discard("F")
print(f"Discard Set: {name_set}")

# 4. Popping
name_set.pop()
print(f"Popping Set: {name_set}")
```

5. Union

```python
num_set_1 = {1,2,3}
num_set_2 = {3,4,5}

result = num_set_1.union(num_set_2)
print("Result of Union Set: ",result)
```
